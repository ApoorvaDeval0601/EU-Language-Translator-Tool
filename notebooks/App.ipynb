```python
# Define the Gradio interface using a soft theme
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    # Add a title and description using Markdown
    gr.Markdown(
        """
        # Language ID & Translation Tool
        This tool identifies the language (Text or Image) and then uses a
        Helsinki-NLP model from Hugging Face to translate it to English.
        """
    )

    # Create tabs for different input types
    with gr.Tabs():
        # Define the "Text Input" tab
        with gr.TabItem("Text Input") as text_tab:
            text_in = gr.Textbox(lines=7, label="Type Text Here", placeholder="Bonjour, comment ça va?")
        # Define the "Image Input" tab
        with gr.TabItem("Image Input") as image_tab:
            image_in = gr.Image(type="numpy", label="Upload an Image with Text")

    # Add a primary submit button
    submit_btn = gr.Button("Process Input", variant="primary")

    # Add a horizontal rule and a "Results" heading
    gr.Markdown("---")
    gr.Markdown("### Results")

    # Add an output text box for the identified language
    lang_out = gr.Textbox(label="Identified Language", interactive=False, lines=1)

    # Create a row for displaying detected and translated text
    with gr.Row():
        processed_text_out = gr.Textbox(label="Text Detected from Source", interactive=False, lines=8)
        trans_out = gr.Textbox(label="English Translation", interactive=False, lines=8)

    # Connect the submit button to the main processing function
    submit_btn.click(
        fn=identify_and_translate,
        inputs=[text_in, image_in],
        outputs=[lang_out, trans_out, processed_text_out]
    )

    # Define event handler for the "Text Input" tab
    # This clears the image input when the text tab is selected
    text_tab.select(
        fn=lambda: None,
        inputs=None,
        outputs=[image_in]
    )

    # Define event handler for the "Image Input" tab
    # This clears the text input when the image tab is selected
    image_tab.select(
        fn=lambda: None,
        inputs=None,
        outputs=[text_in]
    )

# Standard Python entry point
if __name__ == "__main__":
    # Check if the language detection model (pipe_lr) is loaded
    if pipe_lr is None:
        print("\n--- WARNING: GRADIO WILL LAUNCH, BUT PREDICTIONS WILL FAIL ---")
        print("--- Please ensure 'language_detection_model.joblib' is in the correct folder. ---")

    # Launch the Gradio application with a public share link
    print("\nLaunching Gradio UI...")
    demo.launch(share=True)
```
